```
#Usando BS para scrapear data. Ejemplo con la Main Page de Rava.com
#tuto de YT: https://www.youtube.com/watch?v=Cb_5A6geOUw&t=575s
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```
lista_especies = []
lista_ultimo = []
lista_dia = []
lista_mes = []
lista_a√±o = []
```

```
URL = "http://www.rava.com/"
page = requests.get(URL)
soup = BeautifulSoup(page.text, "html.parser")
especies = soup.find_all("tr", attrs={"class": "ci"})
movimientos = soup.find_all("td", attrs={"class": "c1"})
```

```
# soup
```

```
especies
```

```
#loops

for i in especies :
  row = i.find_all("td")
  lista_especies.append(row[0].text.strip()) #Ticker
  lista_ultimo.append(row[1].text.strip()) #Ultimo
  lista_dia.append(row[2].text.strip()) #dia 
```

```
# lista_ultimo
```

```
lista_especies
```

```
#Crear el df
data = pd.DataFrame()
data["especie"] = lista_especies
data["ultimo"] = lista_ultimo
data["dia"] = lista_dia
data = data[data["especie"] != ""]

data
```
